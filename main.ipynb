```python
# I manually scrolled through my messages to see the last time I dumped my messages into notes
year = 2020
month = 2
day = 16
```

```python
#we now want to query messages WHERE handle_id = 19
mymessages = pd.read_sql_query(
    f"""
    SELECT 
        text
        , strftime('%Y', datetime(date/1000000000 + strftime("%s", "2001-01-01") ,"unixepoch","localtime")) as year
        , strftime('%m', datetime(date/1000000000 + strftime("%s", "2001-01-01") ,"unixepoch","localtime")) as month
        , strftime('%d', datetime(date/1000000000 + strftime("%s", "2001-01-01") ,"unixepoch","localtime")) as day
        , ROWID
    FROM message
    WHERE handle_id = 19
        AND year > {year}
        AND month > {month}
        AND day > {day}
    ORDER BY 
        year DESC
        , month DESC
        , day DESC
    LIMIT 300
    """, conn, index_col='ROWID')
```

```python
#you get duplicate messages when you text yourself! (and DISTINCT did not filter correctly)
mymessages.drop_duplicates(subset="text", inplace=True)
```
